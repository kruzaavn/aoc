```python
def check_password(example):
    
    conditions, password = example.split(":")
    
    low_high, value = conditions.split()
    
    low, high = list(map(int, low_high.split('-')))
    
    value_in = sum(map(lambda x: value in x, password))
    
    return low <= value_in <= high
    
```

```python
sum(map(check_password, data))
```

```python
def check_index_password(example):
    
    conditions, password = example.split(":")
    
    password = password.strip()
    
    low_high, value = conditions.split()
    
    low, high = list(map(int, low_high.split('-')))
    
    return (password[low - 1] == value) ^ (password[high - 1] == value)
    
```

```python
check_index_password('1-3 b: cdefg')
```

```python
sum(map(check_index_password, data))
```
